```python
##q1:Durgunluk dönemi öncesi ve sonrası (2008 yılı), İşsizlik oranı her ülkede nasıl değişti?
q1= data[['Country Name','2007','2008','2009']]
q1
```

```python
q1[q1['Country Name']=='Turkiye']
```

```python
plt.rcParams["figure.figsize"] = (10,6)
q1[q1['Country Name']=='Turkiye'].plot.bar(title='Unemployement rate in Turkiye during 2007-2009')
plt.ylabel('Unemployement rate')
plt.xlabel('Turkiye')
plt.show()
```

```python
#q2:Pandemi dönemi öncesi ve sırasında, her ülkede işsizlik nasıl değişti?
q2 =data[['Country Name','2019','2020','2021']]
q2
```

```python
plt.rcParams["figure.figsize"] = (10,6)
q2[q2['Country Name']=='Turkiye'].plot.bar(x='Country Name',title='Unemployment rate for Turkiye during the Pandemic period')
plt.ylabel('Unemployement rate')
plt.xlabel('Country Name')
plt.show()
```

```python
Turkiye_data = data.loc[data['Country Name'] == 'Turkiye']
Turkiye_data
```

```python
#1991 ile 2021 yılları arasında Türkiye'de maksimum ve minimum işsizlik oranı
TUR_max_min = Turkiye_data.melt(id_vars=['Country Name', 'Country Code'], var_name='Years', value_name='Rate %')
TUR_max_min
```

```python
sns.lineplot(data=TUR_max_min, x='Years', y='Rate %');
plt.xticks(rotation=60);#buradaki kod yıllar kısmındaki indexlerin düzenini belirlemek için var
plt.title('Unemployment rate in Turkiye between 1991 and 2021')
```
